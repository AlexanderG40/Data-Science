```python
# import beautiful soup
from bs4 import BeautifulSoup
# parse through HTML and XML files, maybe even Json files
```

```python
import requests

quotes_to_scrape = 'https://quotes.toscrape.com/'

Parsed_Data = requests.get(quotes_to_scrape)
#Parsed_Data

data = Parsed_Data.text
data
```

```python
# save it into a file to read it
with open('quotes.html', 'w') as f:
    f.write(data)
```

```python
# here we finally use beautiful soup
data_Bs = BeautifulSoup(data, 'html.parser')
data_Bs
```

```python
# we give a name to make it easier to find
top_Ten_quotes = data_Bs.find_all("a")
top_Ten_quotes
```

```python
select_Class = 'tag-item'
# create a variable called top ten tags
top_Ten_tags = data_Bs.find_all('span', {'class': select_Class})
top_Ten_tags
```

```python
# create an empty list and call it categories
categories = []
# create a for loop for our tags
for tag in top_Ten_tags:
  category_tag_link = tag.find('a')
  # run a if statement
  if category_tag_link:
    category = category_tag_link.text.strip()
    categories.append(category)
categories
```

```python
# we will find all the diff tags with the class name of 'quote'
quote_elements = data_Bs.find_all("div", class_ = "quote")
# quote_Elements
# returns all the div tags

# we use a method to extract three different things

quotes = []
authors = []
categories_for_quotes = []

for quote_element in quote_elements:
  quote_text = quote_element.span.text
  author_name = quote_element.find("small", class_ = "author").text
  category_name = quote_element.find("a", class_ = "tag").text
# we append the info into our empty lists
  quotes.append(quote_text)
  authors.append(author_name)
  categories_for_quotes.append(category_name)
```

```python
# we save this into a data frame
```

```python
import pandas as pd

final_df ={
    'Top Ten Tags': categories,
    'Quotes': quotes,
    'Authors': authors,
    'Categories': categories_for_quotes
}

df = pd.DataFrame(final_df)
df
```

```python
# we do DF size to get the size of our data
df.size
```

```python
# here we will search for the most common category
# according to https://datascientyst.com/get-most-frequent-values-pandas-dataframe/
# .mode will get us the most frequent value in a column which works perfectly
df['Categories'].mode()
```

So, using this we know that 'misattributed-eleanor-roosevelt' appears the most and is the most frequent with a value of 9.
