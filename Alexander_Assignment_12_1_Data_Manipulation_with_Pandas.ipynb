```python
# we import our pandas for this assignment
import pandas as pd
# import our dataframe
df = pd.read_csv("/content/Fast-Food Chains in USA.csv")
df.sample(25)
```

Write a Python code snippet using pandas to insert a new record into the dataset with the following details: Fast-Food Chain: "Subway", Systemwide Sales: $1000 million, Average Sales per Unit: $400 thousand, Franchised Stores: 2000, Company Stores: 500, 2021 Total Units: 2500, Total Change in Units from 2020: 50

```python
# we insert our new data into the data frame
# researched and found out we can add a new row using the .loc https://www.geeksforgeeks.org/how-to-add-one-row-in-an-existing-pandas-dataframe/
df.loc[len(df)] = ["Subway", 1000, 400, 2000, 500, 2500, 50]
df.tail()
```

Write a Python code snippet using pandas to update the systemwide sales for the fast food chain with the name "McDonald's" to $40000 million.

```python
# we use the .loc feature to update our sales
# https://dnmtechs.com/updating-values-in-a-specific-row-in-a-python-pandas-dataframe/
# it searches it based as a boolean it only changes it where it finds "McDonalds" in its name

df.loc[df['Fast-Food Chains'] == 'McDonald’s', 'U.S. Systemwide Sales (Millions - U.S Dollars)'] = 40000
df
```

 Write a Python code snippet using pandas to retrieve the top 10 fast food chains with the highest average sales per unit, ordered by average sales per unit in descending order.

```python
# we used the .nlargest which automatically gets the top 10 or whatever and puts them in descdending order automatically
# https://datascientyst.com/get-top-10-highest-lowest-values-pandas/
df["Average Sales per Unit (Thousands - U.S Dollars)"].nlargest(n=10)
```

Write a Python code snippet using pandas to retrieve the names of all fast food chains along with their total units for 2021.

```python
# https://www.geeksforgeeks.org/how-to-select-multiple-columns-in-a-pandas-dataframe/
# this was quite simple since we just needed to get the data from two full columns we say what columns we want to retrive data and thats it
df[['Fast-Food Chains', '2021 Total Units']]
```

Write a Python code snippet using pandas to retrieve the names of all fast food chains that have more franchised stores than company stores.

```python

# Again while reasearching I found it to be more simple than I thought we just use operators to say if something is greater than we print it
# https://www.geeksforgeeks.org/ways-to-filter-pandas-dataframe-by-column-values/
fastfood_df = df[df['Franchised Stores'] > df['Company Stores']]

fastfood_df
```

Write a Python code snippet using pandas to delete all records from the dataset where the total change in units from 2020 is negative.

```python
# we use the df.drop
negative_values = df[df['Total Change in Units from 2020'] < 0].index
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html
# https://www.geeksforgeeks.org/how-to-drop-rows-in-dataframe-by-conditions-on-column-values/
# we use the df.drop
df = df.drop(negative_values)

negative_values
```
